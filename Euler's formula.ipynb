```python
notes:
e**(i*k) = cos(k) + i*sin(k)

modified magnitude

M*(e**(i*k)) = M*(cos(k) + sin(k)) ----where M = magnitude

examples:
    e**(i*0) = 1 + 0**i ------ = [1,0]
    
    e**(i*pi/2) = 0 + 1*i ------ = [0,1]
    
    eulers formula :: 
    e**(i*pi) = -1 + 0*i ------- = [-1,0] (also e**(i*pi) + 1 = 0)
    contains all basic mathematical elements
    
    
    
```

```python
# --these values are also defined in above cell
k = np.pi/6
m = 2.3
eul = m*np.exp(1j*k)
# ---

mag = np.abs(eul)
ang = np.angle(eul)

plt.polar([0,ang],[0,mag],'r-')
plt.polar(k,m,'bo')

plt.show()
```

```python
# exercise.independent
def eulerFromCosSin():
    
    # get inputs from user
    sin_ = int(input('Enter sine value?'))
    cos_ = int(input('Enter cosine value'))
    eul = (np.cos(cos_)) + (1j*np.sin(sin_))
    
    # plot
    print(eul)
    plt.polar([0,np.real(eul)],[0,np.imag(eul)],'r-')
    plt.polar(np.real(eul),np.imag(eul),'go')
    
    plt.title('')
    
    
eulerFromCosSin()
```
