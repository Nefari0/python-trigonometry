```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# uniform distribution
size = 100
random = np.random.rand(size)
x = np.linspace(random,random,size)
plt.plot(x,random)
plt.gca().set_aspect('equal')
plt.show()

```

```python
# normal distribution
size = 100
r = size
randomn = np.random.randn(size)
x = np.linspace(randomn,randomn,size)
plt.plot(x,randomn)
plt.gca().set_aspect('equal')
plt.show()
```

```python

angs = np.random.rand(100)*2*np.pi

for i in angs:
    plt.plot([0,np.cos(i)],[0,np.sin(i)])
    
plt.axis('square')
plt.axis([-1,1,-1,1])
plt.axis('off')
plt.show()

```
