```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# uniform distribution
size = 100
random = np.random.rand(size)
x = np.linspace(random,random,size)
plt.plot(x,random)
plt.gca().set_aspect('equal')
plt.show()

```

```python
# normal distribution
size = 100
r = size
randomn = np.random.randn(size)
x = np.linspace(randomn,randomn,size)
plt.plot(x,randomn)
plt.gca().set_aspect('equal')
plt.show()
```

```python
# exercise.independent
size = 100 # demensions
x = np.linspace(0,2*np.pi,100) # circle
phs = np.random.rand(size)*2*np.pi # angles


# generate the plot
plt.plot(np.cos(x),np.sin(x),'k')



for i in range(size):
    plt.plot([0,np.cos(phs[i])],[0,np.sin(phs[i])],'r-')
# plt.plot([0,np.cos(phs)],[0,np.sin(phs)],'r-')
plt.gca().set_aspect('equal')
plt.show()
```

```python
# exercise.instructed
angs = np.random.rand(100)*2*np.pi

for i in angs:
    plt.plot([0,np.cos(i)],[0,np.sin(i)])
    
plt.axis('square')
plt.axis([-1,1,-1,1])
plt.axis('off')
plt.show()

```
