```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# uniform distribution
nums = np.random.rand(1000) 
# nums = nums + 3 # shifts distribution

# generate desired min and max values-------
minval = 2
maxval = 17
nums = nums*(maxval-minval) + minval

plt.plot(nums,'ro') # values on y axis
plt.show()

plt.hist(nums) # values on x axis
plt.show()
```

```python
# normal distribution
nums = np.random.randn(1000) # n in rand = normal

plt.plot(nums,'ro',alpha=.5) # values on y axis # alpha changes transparency
plt.show()

plt.hist(nums) # values on x axis
plt.show()
```
