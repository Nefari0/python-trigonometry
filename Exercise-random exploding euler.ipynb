```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# angs = np.random.rand(100)*2*np.pi
k = np.linspace(0,np.pi,100)
# m = 2.3
m = list(np.random.randn(100))

pnt = 6

eul = m*np.exp(1j*k)

# ---cis for cos, imaginary and sin---
cis = m*(   np.cos(k) + 1j*np.sin(k)   )

# print(np.imag(cis[pnt]),np.real(cis[pnt]))
for i in range(len(k)):
    plt.plot([0,np.imag(cis[i])],[0,np.real(cis[i])])
plt.axis('off')
plt.axis([-1.2,1.2,-1.2,1.2])
```

```python
# exercise.instructed
nvects = 200
ms = np.random.rand(nvects)
ks = np.random.rand(nvects)*2*np.pi

for i in range(0,nvects):
    
    r = np.random.rand()
    
    if r<.4:
        clr = [1,.2,.7] #color#pink = 1 red .2 green .7 blue
    elif r>.4 and r<.8:
        clr = [.7,.2,1] #color#purple = .7red .2green 1blued
    else:
        clr = [0,1,0]
    
    plt.polar([0,ks[i]],[0,ms[i]],color=clr)

plt.axis('off')
plt.thetagrids([])
plt.show()
```
